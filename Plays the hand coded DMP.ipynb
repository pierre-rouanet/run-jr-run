```python
import time

for _ in range(2):
    env.recorder.start()
    env.dmp_player.play(env.base_dmp, 
                        duration=.6, repeat=10, 
                        wait=True)
    env.recorder.stop()
    time.sleep(1)
    env.drive_back()
```
